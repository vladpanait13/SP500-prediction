```python
import yfinance as yf
```

```python
sp500 = yf.Ticker("^GSPC")
```

```python
sp500 = sp500.history(period = "max")
```

```python
sp500.plot.line(y = "Close", use_index = True)
```

```python
del sp500 ["Dividends"]
del sp500 ["Stock Splits"]
```

```python
sp500.head()
```

```python
sp500["Tomorrow"] = sp500["Close"].shift(-1)
sp500
```

```python
# This returns a boolean value but we convert it into an int in order to use it in our Machine Learning model

sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
```

```python
# Our new Target variable: 0 - price went down, 1 - price went up (today price > yesterday price)
sp500
```

```python
# Remove data older than 1990 
sp500 = sp500.loc["1990-01-01":].copy()
```

```python
sp500
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# in stock market, there are non-linear relationships between the open price and the target, making Random Forest a good algorithm
# n estimators = number of individual decision trees we want to train. The higher this is, the better the accuracy is.
# min_samples_split = protect against overfitting. the higher is it, the less the accuracy is, but it doesn't overfit
# random_state = a random forest has randomization built-in. setting a random state means if we run the model twice, 
#the random numbers generated will be in a predictable sequence each time using this random seed of 1, meaning we get same results

model = RandomForestClassifier(n_estimators = 100, min_samples_split = 100, random_state = 1)

# split the data by putting all the rows except the last hundred rows into the training set. we put the last hundred in the test.
train = sp500.iloc[:-100]
test = sp500.iloc[-100:]

predictors = ["Close", "Volume", "Open", "High", "Low"]

# fit the model using the predictors and try to predict the target
model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score 

preds = model.predict(test[predictors])
```

```python
import pandas as pd

preds = pd.Series(preds, index = test.index)
```

```python
precision_score(test["Target"], preds)
```

```python
combined = pd.concat([test["Target"], preds], axis = 1)
```

```python
combined.plot()
```

```python
# backtesting using a predict function

def predict(train, test, predictors, model):
    model.fit(train[predictors], train ["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index = test.index, name = "Predictions")
    combined = pd.concat([test["Target"], preds], axis = 1)
    return combined
```

```python
def backtest(data, model, predictors, start = 2500, step = 250):
    all_predictions = []
    
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
# percentage accuracy - 52.8%

precision_score(predictions["Target"], predictions["Predictions"])
```

```python
# percentage of days when the market went up

predictions["Target"].value_counts() / predictions.shape[0]
```

```python
# the last closing price between 2, 5, 60, 250 and 1000 days

horizons = [2, 5, 60, 250, 1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]
    
    trend_column = f"Trend {horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors += [ratio_column, trend_column]
    
```

```python
sp500 = sp500.dropna()
sp500
```

```python
model = RandomForestClassifier(n_estimators = 200, min_samples_split = 50, random_state = 1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pd.Series(preds, index = test.index, name = "Predictions")
    combined = pd.concat([test["Target"], preds], axis = 1)
    return combined
```

```python
predictions = backtest(sp500, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
